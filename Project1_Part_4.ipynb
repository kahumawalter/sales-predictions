```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

filename = ('/content/drive/MyDrive/Coding Dojo/Week 1 /sales_predictions.csv')
df = pd.read_csv(filename)

df.head()
```

```python
df.info()
```

```python
#The boxplot visualization belows show the item outlet sales of the different Items sold. The interquartile range (IQR) of most items is between 1800 and 3500


sns.boxplot(x='Item_Type', y='Item_Outlet_Sales',data=df);
plt.xticks(rotation = 90);
plt.title('Sales of Different Items', fontsize =16);

#This tells Item_weight of the different Item_Types
```

```python
# The graph below shows the item sales in comparison item weight


df2 = pd.DataFrame(np.random.rand(10, 2), columns=['Item_Weight', 'Item_Outlet_Sales'])
df2.plot(stacked=True)
plt.xlabel('Item_Weight', fontsize = 15)
plt.ylabel('Item_Outlet_Sales', fontsize = 15)
plt.title('Item Sales in Comparison To Item Weight', fontsize = 25);
```
