```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

filename = ('/content/drive/MyDrive/Coding Dojo/Week 1 /sales_predictions.csv')
df = pd.read_csv(filename)

df.head()
```

```python
df['Item_Outlet_Sales'].hist();
```

```python
df['Item_Outlet_Sales'].hist(bins = 10)
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
# finding outlet sales less than or equal to 3000

plt.style.use('seaborn')
price_filter = df.loc[:, 'Item_Outlet_Sales'] <= 3000
df.loc[price_filter, 'Item_Outlet_Sales'].hist(bins = 30,
                                   edgecolor='black')
plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
```

```python
#Using the seaborn wrapper to get summaries of 'Item_Type' and 'Item_Weight'
sns.boxplot(x='Item_Type', y='Item_Weight',data=df);
```

```python
sns.boxplot(x='Item_Type', y='Item_Weight',data=df);
plt.xticks(rotation = 90);

#This tells Item_weight of the different Item_Types
```

```python
#Heatmap of the correlation between features.

corr = df.corr()

import seaborn as sns
sns.heatmap(corr, cmap = 'Blues')
```

```python
sns.heatmap(corr, cmap = 'Blues', annot = True)
```
